```python
import numpy as np
import matplotlib.pyplot as plt
import mpmath as mp
PI = np.pi
```

## 例題

### 2-1

```python
def partial_sum(t, n):
    s = PI / 2
    i = 1
    while i <= n:
        s -= 4/PI * (np.cos(i*t) / i**2)
        i += 2
    return s
t = np.linspace(-PI, PI, 1000)
x = np.abs(t)
s1 = partial_sum(t, 1)
s2 = partial_sum(t, 3)
plt.plot(t, x)
plt.plot(t, s1)
plt.plot(t, s2)
plt.show()
```

```python
I = [-PI, PI]
x = lambda t: abs(t)
ps = mp.fourier(x, I, 3)
psval = lambda t: mp.fourierval(ps, I, t)
mp.plot([x, psval], xlim=I)
```

### 2-3

```python
def partial_sum(t, n):
    s = 0
    i = 1
    while i <= n:
        s += np.sin(i*t) / i
        i += 2
    return 4 / PI * s

def clk(t):
    if t < 0:
        return -1
    else:
        return 1
    
t = np.linspace(-PI, PI, 1000)
x = np.vectorize(clk)(t)
s = partial_sum(t, 3)
plt.plot(t, x)
plt.plot(t, s)
plt.show()
```

### 2-4

```python
def partial_sum_cos(t, n):
    s = PI / 2
    i = 1
    while i <= n:
        s -= 4/PI * np.cos(i*t) / (i**2)
        i += 2
    return s

def partial_sum_sin(t, n):
    s = 0
    for i in range(1, n+1):
        sign = 1
        if i % 2 == 0:
            sign = -1
        s += sign * 1/i * np.sin(i*t)
    return 2 * s

def saw(t):
    return t

t = np.linspace(0, PI, 10000)
x = np.vectorize(saw)(t)
s_cos = partial_sum_cos(t, 3)
s_sin = partial_sum_sin(t, 30)
plt.plot(t, x)
plt.plot(t, s_cos)
plt.plot(t, s_sin)
plt.show()
```

```python
def partial_sum(t, n):
    s = 0
    i = 1
    sign = 1
    while i <= n:
        v = 2 * PI * sign / i
        if i % 2 == 1:
            v -= 8 / (i**3 * PI)
        v *= np.sin(i * t)
        s += v
        sign *= -1
        i += 1
    return s


def f(t):
    return t*t

t = np.linspace(0, PI, 10000)
x = np.vectorize(f)(t)
s = partial_sum(t, 20)
plt.plot(t, x)
plt.plot(t, s)
plt.show()
```
