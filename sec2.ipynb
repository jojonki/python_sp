```python
import numpy as np
import matplotlib.pyplot as plt
PI = np.pi
```

## 例題

### 2-1

```python
def partial_sum(t, n):
    s = PI / 2
    i = 1
    while i <= n:
        s -= 4/PI * (np.cos(i*t) / i**2)
        i += 2
    return s
t = np.linspace(-PI, PI, 1000)
x = np.abs(t)
s1 = partial_sum(t, 1)
s2 = partial_sum(t, 3)
plt.plot(t, x)
plt.plot(t, s1)
plt.plot(t, s2)
plt.show()
```
