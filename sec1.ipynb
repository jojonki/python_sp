```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1- 1

```python
x = np.array([-3, 2, 8, 9])
def step_func_1_1(k):
    return x[k]**3 - 3*(x[k]**2) + 5*x[k] + 2
np_step_func_1_1 = np.vectorize(step_func_1_1)
# k = np.linspace(0, 3, 4).astype(np.int64)
k = np.arange(0, 4)
print(np_step_func_1_1(k))
```

### 1-2

```python
x = np.array([-2, 3, 5, 7])
def step_func_1_1(k):
    return np.exp(-x[k]**2)
np_step_func_1_1 = np.vectorize(step_func_1_1)
k = np.arange(0, 4)
print(np_step_func_1_1(k))
```

### 1-3

```python
z = np.array([-1.0+2.0j, 2.3+3.5j, -3.7+0.1j])
print('z.real:', z.real)
print('z.imag:', z.imag)
exp_z = np.exp(z.real) * (np.cos(z.imag) + 1j*np.sin(z.imag))
print('exp(z)1:', np.exp(z))
print('exp(z)2:', exp_z)
```

### 1-4

```python
# x(t)=exp(-|t|)cos(30t)
# x(t)=log(1+t)
t1 = np.linspace(-5, 5, 100)
x1 = np.exp(-np.abs(t1)) * np.cos(30.0*t1)

t2 = np.linspace(0, 5, 100)
x2 = np.log(1+t2)

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(t1, x1)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(t2, x2)
plt.show()
```

### 1-7

```python
def step_func(t):
    if t >= 1:
        return np.exp(-(t-1)**2)
    elif 0 <= t < 1:
        return 1
    else:
        return np.cos(10.0 * t)
step_func = np.vectorize(step_func)
t = np.linspace(-2, 6, 100)
x = step_func(t)
plt.plot(t, x)
plt.show()
```
