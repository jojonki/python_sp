```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import eval_legendre, eval_chebyc
import sympy as sym
```

### 3.1

```python
PI = sym.S.Pi

t = sym.Symbol('t')
I = sym.integrate((sym.sin(t) - sym.cos(t))**2, (t, -PI, PI))

print(I)
```

### 3.2

```python
for j in range(6):
    print(sym.legendre(j, t))
```

### 3.3

```python
t = np.linspace(-1, 1, 256)
legen  = eval_legendre(20, t)
plt.plot(t, legen)
plt.show()
```

```python
t = sym.Symbol('t')
I = sym.integrate((sym.sin(t) - sym.cos(t)**3)**2, (t, -PI, PI))
print(I) # d(x1, x2)
J = sym.integrate((sym.cos(t)**3)**2, (t, -PI, PI))
print(J) 
```
